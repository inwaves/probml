```python
pv = 0.5 * np.ones(M)   # prior variance
w = np.zeros(M)         # skills, initialized to be the prior mean μ0 = 0
skills_across_iterations = [w]

for iteration in tqdm(range(1100)):
    # Sample performance differences (t) given skills (w) and outcomes (G)
    s = w[G[:,0]] - w[G[:,1]] # s is the skill difference.
    σ = 1
    
    # Sample t according to the skill difference plus some noise epsilon.
    t = s + σ * scipy.stats.norm.ppf(1 - np.random.uniform(size=N)*(1-scipy.stats.norm.cdf(-s/σ)))

    # Sample skills given performance differences
    inverse_Sigma_tilde = np.zeros((M, M))
    mu_tilde = np.zeros(M)
    
    # Compute Σ~^{-1}
    for g in range(N):
        inverse_Sigma_tilde[G[g, 0], G[g, 0]] += 1
        inverse_Sigma_tilde[G[g, 1], G[g, 1]] += 1
        inverse_Sigma_tilde[G[g, 0], G[g, 1]] -= 1
        inverse_Sigma_tilde[G[g, 1], G[g, 0]] -= 1
        
    # Compute μ~
    for i in range(M):
        for g in range(N):
            if G[g, 0] == i:
                mu_tilde[i] += t[g]
            elif G[g, 1] == i:
                mu_tilde[i] -= t[g]
        
    Σinv = np.diag(1/pv) + inverse_Sigma_tilde
    Σ = np.linalg.inv(Σinv)
    μ = Σ @ mu_tilde
    w = np.random.multivariate_normal(mean=μ, cov=Σ)
    
    # Storing results across iterations for plotting.
    skills_across_iterations.append(w)
```

```python
def gaussian_ep(G, M):
    def Ψ(x): return scipy.stats.norm.pdf(x) / scipy.stats.norm.cdf(x)
    def Λ(x): return Ψ(x) * (Ψ(x) + x)
    N = len(G)

    μ_s, p_s = np.empty(M), np.empty(M)
    μ_gs, p_gs = np.zeros((N,2)), np.zeros((N,2))
    μ_sg, p_sg = np.empty((N,2)), np.empty((N,2))
    
    while True:
        # 1. Compute marginal skills
        # Let skills be N(μ_s, 1/p_s)
        p_s = np.ones(M) * 1/0.5
        μ_s = np.zeros(M)
        for j,(winner,loser) in enumerate(G):
            p_s[winner] += p_gs[j,0]
            p_s[loser] += p_gs[j,1]
            μ_s[winner] += μ_gs[j,0] * p_gs[j,0]
            μ_s[loser] += μ_gs[j,1] * p_gs[j,1]
        μ_s = μ_s / p_s

        # 2. Compute skill -> game messages
        # winner's skill -> game: N(μ_sg[,0], 1/p_sg[,0])
        # loser's skill -> game: N(μ_sg[,1], 1/p_sg[,1])
        p_sg = p_s[G] - p_gs
        μ_sg = (p_s[G]*μ_s[G] - p_gs*μ_gs) / p_sg

        # 3. Compute game -> performance messages
        v_gt = 1 + np.sum(1/p_sg, 1)
        σ_gt = np.sqrt(v_gt)
        μ_gt = μ_sg[:,0] - μ_sg[:,1]

        # 4. Approximate the marginal on performance differences
        μ_t = μ_gt + σ_gt * Ψ(μ_gt/σ_gt)
        p_t = 1 / v_gt / (1-Λ(μ_gt/σ_gt))

        # 5. Compute performance -> game messages
        p_tg = p_t - 1/v_gt
        μ_tg = (μ_t*p_t - μ_gt/v_gt) / p_tg

        # 6. Compute game -> skills messages
        # game -> winner's skill: N(μ_gs[,0], 1/p_gs[,0])
        # game -> loser's skill: N(μ_gs[,1], 1/p_gs[,1])
        p_gs[:,0] = 1 / (1 + 1/p_tg + 1/p_sg[:,1])  # winners
        p_gs[:,1] = 1 / (1 + 1/p_tg + 1/p_sg[:,0])  # losers
        μ_gs[:,0] = μ_sg[:,1] + μ_tg
        μ_gs[:,1] = μ_sg[:,0] - μ_tg
        
        yield (μ_s, np.sqrt(1/p_s))
```

1. To estimate $\theta$, we would have to sample from a posterior distribution $p(\theta | \lambda, \mu, y)$. This would be another discrete distribution over the years $\{1851, \dots, 1961\}$.
2. To get this entire distribution, because $\theta$ is countable, we could look at each year $k$ and find $Pr(\theta = k | \lambda, \mu, y)$, for $k \in \{1851, \dots, 1961\}$.
3. To get the above probability, we can apply Bayes' rule for each $\theta = k$:
    $$ Pr(\theta = k | \lambda, \mu, y) = \frac{Pr(\lambda, \mu, y | \theta = k)\cdot Pr(\theta = k)}{Pr(\lambda, \mu, y)} \propto Pr(\lambda, \mu, y | \theta = k)$$
    The proportionality holds because the prior for $\theta$ is a constant and so is the denominator. We can just say that the posterior distribution is equal to some constant times the likelihood, and then ignore it altogether, since it is not what matters here.
4. We want to find $Pr(\lambda, \mu, y | \theta = k)$.
    - $\lambda, \mu$'s priors are independent from $\theta$, but $y$ isn't.
    - we can marginalise the two parameters $\lambda, \mu$, and use Monte-Carlo sampling to approximate the integral.
    $$\begin{aligned}
         Pr(\lambda, \mu, y | \theta = k) &= \iint Pr(y | \theta = k, \lambda, \mu)Pr(\lambda)Pr(\mu) d \lambda d\mu  \\ 
                                                           &= \iint \text{Poisson}(\lambda \mathbb{1}_{i \leq \theta} + \mu \mathbb{1}_{i > \theta}) \cdot \Gamma(a, b)^2 d \lambda d \mu \\
                                                           & \approx \frac{1}{T} \sum_{t=1}^T Pr(y | \theta = k, \lambda = \lambda_t, \mu = \mu_t)
        \end{aligned}$$
5. We repeat step 4 for all the values $k$ that $\theta$ can take. That gives us the entire distribution $p(\theta | \lambda, \mu, y)$. 
6. In a loop, we sample a $\theta$ from the distribution in step 5., and then calculate $p(\lambda, \mu, y | \theta)$. Then, we sample $\theta$ again, given $p(\theta | \lambda, \mu, y)$.
